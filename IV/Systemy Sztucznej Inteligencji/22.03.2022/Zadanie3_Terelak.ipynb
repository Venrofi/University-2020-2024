```python
asortyment = pd.read_csv('C:\\Users\\User\\Desktop\\Systemy Sztucznej Inteligencji\\22.03.2022\\asortyment_Terelak.csv')
asortyment.head(10)
```

```python
asortyment = asortyment.values.tolist()
```

```python
def find(arg, wages, names):
    data = []
    for i in range(len(arg)):
        result = 0
        for j in range(len(arg[i])):
            result += arg[i][j] * wages[j]
        data.append(result)
    return names[data.index(max(data))]
```

```python
wagi = {"świeże": 1, "mrożone": 0.3, "ostre": 0.1, "słodkie": 0.5, "zielone": 0.5, "czerwone": 0.5,
        "lokalne": 1, "tropikalne": 0.3, "liściaste": 0.5, "bulwowe": 0.5}
wagi = list(wagi.values())
col_names = []

for i in range(len(asortyment)):
    col_names.append(asortyment[i][0])
    del asortyment[i][0]

print("Wynik na podstawie przedstawionych parametrów:", find(asortyment, wagi, col_names))
```
