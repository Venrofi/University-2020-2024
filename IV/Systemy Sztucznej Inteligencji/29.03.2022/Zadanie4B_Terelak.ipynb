```python
class NaiveBayes:
    @staticmethod
    def srednia(atrybut): # Średnia arytmetyczna
        return sum(atrybut) / len(atrybut)
    
    @staticmethod
    def klasyfikator_Skalar(data, sample): # Inna funkcja przynależności - singleton, poprzednio Gauss
        P = {'Setosa': 0, 'Versicolor': 0, 'Virginica': 0}
        
        for klasa in ['Setosa', 'Versicolor', 'Virginica']:
            # Separacja na klasy, usunięcie ostatniej kolumny
            data_set = data[data['variety'] == klasa]
            data_set = data_set.drop(['variety'], axis = 1)
            #print(klasa + ":\n", data_set)
            
            sample_index = 0 # indeks obecnie sprawdzanego atrybutu
            
            for atrybut in ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']:
                sr = NaiveBayes.srednia(data_set[atrybut])
                if(sample[sample_index] > sr): # Jeśli wartość atrybutu próbki jest większa od średniej mnóż razy 1
                    P[klasa] += 1 * sample[sample_index]
                else:
                    P[klasa] += 0 * sample[sample_index]
                
                sample_index += 1
                
        maks = max(P, key = P.get)
        print("Wynik klasyfikacji:\n", P,"\nUzyskana odmiana: ", maks)
        return maks
        
```

```python
NaiveBayes.klasyfikator_Skalar(iris, [6.7, 3.1, 5.6, 2.4]) # Virginica
```

```python
NaiveBayes.klasyfikator_Skalar(iris, [5.1, 3.8, 1.9, 0.4]) # Setosa
```

```python
NaiveBayes.klasyfikator_Skalar(iris, [5.7, 2.9, 4.2, 1.3]) # Versicolor
```

```python
test = [[5.7, 2.9, 4.2, 1.3], [5.1, 3.8, 1.9, 0.4], [6.7, 3.1, 5.6, 2.4]]
for i in range(len(test)):
    NaiveBayes.klasyfikator_Skalar(iris, test[i])
```
